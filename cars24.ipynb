```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
car_name=[]
seater=[]
model=[]
fuel_type=[]
available_colors=[]
manual=[]
price=[]
city=[]
image=[]
```

```python
for i in range(0,len(results)):
    car_name.append(results[i].find('h2').get_text())
    fuel_type.append(results[i].find('p',{"data-testid":"car_variant_fuel_type"}).get_text())
    manual.append(results[i].find('p',{"data-testid":"car_variant_transmission"}).get_text())
    price.append(results[i].find('div',{"data-testid":"car_price"}).get_text())
    city.append(results[i].find('div',{"class":"styles__CityName-sc-57d31ed8-17 lhXIXl"}).get_text())
    lt=results[0].find_all('p',{"class":"styles__ParaWithoutMargins-sc-57d31ed8-33 jvhnwy"})
    model.append(lt[0].get_text())
    available_colors.append(lt[5].get_text())
    seater.append(lt[1].get_text())
    image.append(results[i].find('a',{"data-testid":"car_image"})['href'])
    

    
```

```python
colums=
```

```python
cars=pd.DataFrame({"car_name":car_name,
"seater":seater,
"model":model,
"fuel_type":fuel_type,
"available_colors":available_colors,
"manual":manual,
"price":price,
"city":city,
"link":image})
```
